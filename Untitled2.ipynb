```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vacaturesnl = pd.read_csv(r'C:\Users\Menno\Downloads\vacatures.csv', sep = ';')
vacaturesnl.drop(vacaturesnl.columns[[0]], axis=1, inplace=True)
vacaturesnl['vacatures'] = [149.5, 154.8, 162.0, 171.2, 185.7, 204.7, 213.7, 226.5, 234.6, 251.0]
vacaturesnl = vacaturesnl.values
print(vacaturesnl)
```

```python
werklozennl = pd.read_csv(r'C:\Users\Menno\Downloads\werklozen.csv', sep = ';')
periodenw = werklozennl[['periode']]
werklozennl.drop(werklozennl.columns[[0]], axis=1, inplace=True)
werklozennl['werklozen'] = [576.3, 560.7, 524.0, 494.3, 472.0, 452.7, 428.0, 398.7, 368.0, 353.7]
werklozennl = werklozennl.values
periodenw = periodenw.values
print(periodenw)
print(werklozennl) 
```

```python
import seaborn
fig = plt.figure() 
ax = fig.add_subplot(111)
plt.plot(werklozennl, vacaturesnl, linewidth=4.0, color='m')
ax.set(ylabel='Vacatures (duizendtallen)', xlabel='Werklozen (duizendtallen)')
plt.style.use('seaborn-deep')
ax.grid()
plt.show()

```

```python
fig1 = plt.figure() 
ax = fig1.add_subplot(111)
ax2 = ax.twinx()
ax.bar(periodenw, werklozennl, color='lightblue', linewidth=4, label='Werklozen')
ax2.plot(periodenw, vacaturesnl, color='crimson', linewidth = 2, label='Vacatures', marker="o", markersize=8)
plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
ax.legend(loc=9)
ax2.legend(loc=8)
plt.style.use('')
ylabel = ax.set(ylabel='Werklozen (duizendtallen)')
ylabel = ax2.set(ylabel='Vacatures (duizendtallen)')
plt.setp(ylabel, rotation=90)
ax.grid()
plt.show()
```
