```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
```

```python
u = np.linspace(0, 20 , 40 )
v = np.linspace(0, 20, 40)
X,Y = np.meshgrid(u, v)

Z = np.absolute((X**0.5)*(Y**0.5))
```

```python
plt.pcolor(Z)
plt.show()
```

```python
fig1 = plt.figure() 
ax = fig1.add_subplot(111)
plt.contourf(X,Y,Z,20, cmap='viridis')
cbar = plt.colorbar()
cbar.set_label('Utility')
plt.title('Utility Functions')
xlabel = ax.set(xlabel='$Quantity$ $good$ $1$')
ylabel = ax.set(ylabel='$Quantity$ $good$ $2$')
plt.show()
```

```python
x = np.linspace(0, 120 , 10000)
def y(x):
    return ((5-(x**0.5)))**2
x1 = np.linspace(0, 120 , 100)
def x2(x1):
    return 12.5 - x1



```

```python
fig1 = plt.figure() 
ax = fig1.add_subplot(111)
plt.plot(x,y(x),color='green', label='Utility Function', linewidth=2)
plt.plot(x1,x2(x1),color='red', label='Budget Constraint', linewidth=2)
plt.style.use('seaborn-white')
ax.set_xlim(0,15)
ax.set_ylim(0,15)
ax.legend()
ax.grid()
xlabel = ax.set(xlabel='$Quantity$ $good$ $1$')
ylabel = ax.set(ylabel='$Quantity$ $good$ $2$')
plt.annotate('Optimal Bundle', (6.25, 6.25), (6.25+2, 6.25+2),
             arrowprops=dict(facecolor='black') )
rect = patches.Rectangle((-0.1,0),6.25,6.25,linewidth=1,
                         edgecolor='k',facecolor='none', linestyle='--')
ax.add_patch(rect)
plt.annotate('(6.25, 6.25)', (6.25, 6.25), (6.25+0.7, 6.25-0.1)) 
plt.annotate('U=5', (6.5, 14), (6.5, 14))
plt.annotate('Budget=$12.5', (6.5, 13.5), (6.5, 13.5))
plt.title('$Consumer$ $Choice$', size=20)
r = plt.fill_between(x, y(x), y2=0, where=None, 
                 interpolate=False, step=None, hold=None, data=None, color='red', alpha=0.15)
g = plt.fill_between(x, y(x), y2=20, where=None, 
                 interpolate=False, step=None, hold=None, data=None, color='green', alpha=0.15)



plt.show()
```
