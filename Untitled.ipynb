```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bbp = pd.read_csv(r'C:\Users\Menno\Downloads\bbp.csv', sep=';', decimal=',')
bbp['%$\Delta$'] = bbp['bbp']
bbp.drop(bbp.columns[[1]], axis=1, inplace=True)
bbp['jaren'] = bbp['perioden']
bbp.drop(bbp.columns[[0]], axis=1, inplace=True)
jaren = bbp[['jaren']]
bbpnl = bbp[['%$\Delta$']]
jaren= jaren.values
bbpnl = bbpnl.values
print(bbpnl)

```

```python
import seaborn
fig = plt.figure() 
ax = fig.add_subplot(111)
plt.plot(jaren, bbpnl, linewidth=3.0, label='%$\Delta$ BBP')
ax.grid()
ax.set_xticks(jaren, minor=False)
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
ylabel = ax.set(ylabel='%')
plt.style.use('seaborn-deep')
plt.axhline(y=0.0, color='red', alpha=0.5)
ax.legend(loc=1,prop={'size':12})
plt.axvspan(2008, 2010, color='red', alpha=0.5, lw=0)
ax.xaxis.set_tick_params(width=1, length=5)
plt.title('Jaarlijkse groei Nederlandse economie 2000-2017', size=20)
plt.figtext(0.51, 0.8, 'Crisis')





plt.show()
```
